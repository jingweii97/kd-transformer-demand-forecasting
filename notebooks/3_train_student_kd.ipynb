## 3. Verify Teacher Outputs & Soft Targets
Confirm that the precomputed soft target tensor exists in the mounted experiment dataset before training starts.

```python
import os

soft_targets_path = "/kaggle/input/kd-transformer-demand-forecasting-experiments/artifacts/soft_targets/exp_001.pt"
teacher_ckpt_path = "/kaggle/input/kd-transformer-demand-forecasting-experiments/artifacts/teacher/exp_001/best_tft_teacher.ckpt"

if not os.path.exists(soft_targets_path):
    raise FileNotFoundError(f"KD requires soft targets at {soft_targets_path}. Make sure the experiments dataset is correctly mounted.")
if not os.path.exists(teacher_ckpt_path):
    # Check for teacher checkpoint in outputs/ folder structure too
    fallback_teacher = "/kaggle/input/kd-transformer-demand-forecasting-experiments/artifacts/outputs/teacher/exp_001/best_tft_teacher.ckpt"
    if not os.path.exists(fallback_teacher):
        print(f"Warning: Teacher checkpoint not found. Ensure it exists under {teacher_ckpt_path} or fallback.")

print("SUCCESS: Pre-conditions met. Soft targets located.")
```
